```python
list_words = contents.split()
print(len(list_words))
print(list_words)
```

```python
import string
list_words = [w.strip(string.punctuation) for w in list_words]
```

```python
word_freq = {}
for word in list_words:
    word_freq[word] = word_freq.get(word, 0) + 1
```
