```python
def unpack_and_order(result):
  first = result[0]
  second = first['text']
  third = second[2:-2]

  data = eval(third)

  df = pd.DataFrame(columns=['Entity', 'Label'])

  rows = []
  for entry in data:
      if 'E' in entry and 'T' in entry:
          row = {'Entity': entry['E'], 'Label': entry['T']}
          rows.append(row)

  df_filled = pd.concat([df, pd.DataFrame(rows)])
  return df_filled
```
